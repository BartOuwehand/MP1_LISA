```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import h5py
import scipy.signal
#import psd

from lisagwresponse import GalacticBinary
from lisainstrument import Instrument

from pytdi import Data
from pytdi import michelson as mich
```

```python
orbit_path = '../orbits.h5'
gw_path = 'gws.h5'

rec = ['X','Y','Z']
#dur = [500,1000,5000]
#dur = [1000,5000,10000]
dur = [1000,5000,10000,50000,100000]
```

```python
instru = {}
for d in dur:
    instru[d] = Instrument(
        size=d, # in samples
        lock='six',
        orbits=orbit_path, # realistic orbits (make sure it's consistent with glitches and GWs!)
        gws=gw_path
    )
    instru[d].disable_all_noises()
```

```python
# Simulate and write measurements
# !rm measurements/msrmnts500.h5
# !rm measurements/msrmnts1000.h5
# !rm measurements/msrmnts5000.h5
# !rm measurements/msrmnts10000.h5
# !rm measurements/msrmnts50000.h5
# !rm measurements/msrmnts100000.h5


for d in dur:
    instru[d].simulate()
    instru[d].write('measurements/msrmnts' +str(d)+ '_nn.h5')
```

```python
def psd_func(data):
    return scipy.signal.welch(data,fs=4,window='nuttall',nperseg=len(data),detrend=False)
```

```python
# Read data from LISA Instrument
fs = 4
data = {}
for d in dur:
    rawdata = Data.from_instrument('measurements/msrmnts' +str(d)+ '.h5')

    discard = 300
    
    X = mich.X2.build(**rawdata.args)(rawdata.measurements)[discard:]
    #Y = mich.Y2.build(**rawdata.args)(rawdata.measurements)[discard:]
    #Z = mich.Z2.build(**rawdata.args)(rawdata.measurements)[discard:]

    #t = instru[d].t[discard:]
    t = (np.arange(0,len(X)+discard)/4)[discard:]
    
    data[d] = np.array([t,X])

#Create psd for data
psd = {}
for d in dur:
    ftmp, psdtmp = psd_func(data[d][1])
    psd[d] = np.array([ftmp,psdtmp])
```

```python
# Create filtered data and the psd's of this filtered data
fdata = {}
tfdata = {}
coeffs1 = scipy.signal.firls(73,bands=[0,1,1.2,2],desired=[1,1,0,0],fs=4)
coeffs2 = scipy.signal.firls(73,bands=[0,.5,.7,2],desired=[1,1,0,0],fs=4)
for d in dur:
    fdata_tmp = scipy.signal.filtfilt(coeffs1,1., x=data[d][1],padlen=len(psd[d][0]))
    tfdata_tmp = scipy.signal.filtfilt(coeffs2,1.,x=data[d][1],padlen=len(psd[d][0]))

    cutoff = 100
    fdata[d] = np.array([data[d][0][cutoff:-cutoff],fdata_tmp[cutoff:-cutoff]])
    tfdata[d] = np.array([data[d][0][cutoff:-cutoff],tfdata_tmp[cutoff:-cutoff]])

fpsd = {}
tfpsd = {}
for d in dur:
    ftmp, psdtmp = psd_func(fdata[d][1])
    fpsd[d] = np.array([ftmp,psdtmp])
    ftmp, psdtmp = psd_func(tfdata[d][1])
    tfpsd[d] = np.array([ftmp,psdtmp])
```

```python
fig, axs = plt.subplots(1,len(dur))
fig.set_figheight(4)
fig.set_figwidth(4*len(dur))
for i in range(len(dur)):
    d = dur[i]
    axs[i].plot(data[d][0],data[d][1])
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel("Amplitude")
    axs[i].set_title("Data for duration "+str(d))
plt.show()

plt.figure(figsize=(12,8))
for d in dur:
    plt.plot(psd[d][0],psd[d][1],label=str(d))
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Freq [Hz]')
plt.ylabel('ASD [Hz/sqrt(Hz)]?')
plt.title("PSD for different durations")
plt.show()

plt.figure(figsize=(12,8))
for d in dur:
    plt.plot(fpsd[d][0],fpsd[d][1],label=str(d))
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Freq [Hz]')
plt.ylabel('ASD [Hz/sqrt(Hz)]?')
plt.title("PSD for different durations")
plt.show()

plt.figure(figsize=(12,8))
for d in dur:
    plt.plot(tfpsd[d][0],tfpsd[d][1],label=str(d))
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Freq [Hz]')
plt.ylabel('ASD [Hz/sqrt(Hz)]?')
plt.title("PSD for different durations")
plt.show()
```
