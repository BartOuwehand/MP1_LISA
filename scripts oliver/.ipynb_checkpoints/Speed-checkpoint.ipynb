# Why is pyTDI so slow?

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.lines as mlines

import numpy as np
import scipy as sp


import lisagwresponse
import lisaorbits
from lisaconstants import c, au


import pytdi
from pytdi.michelson import X2, Y2, Z2

import os
```

```python

years=4
dt = 8640.
t = np.arange( 0, years*365*86400+dt, dt)
orb_t = np.arange(t[0]-dt, t[-1]+dt, dt )
orbits = lisaorbits.KeplerianOrbits(dt=dt,size=orb_t,t0=t[0]-dt, t =orb_t )
orbits.tau0 = orbits.t0

if os.path.isfile("orbits.h5"):
    os.remove( "orbits.h5")
orbits.write("orbits.h5")
```

```python
gw_dt = 100
gw_t0 = 0
gw_size = int( 4*365*86400/gw_dt  )
```

```python
%%time
gw_galbin = lisagwresponse.GalacticBinary( 1e-18, 1e-3, 
                                          orbits='orbits.h5', gw_lambda=0, gw_beta=0, dt=gw_dt, 
                                          t0=gw_t0, size = gw_size )
```

```python
gw_galbin.version = '1.0'
gw_galbin.write("gw_galbin.h5")
```

```python
%%time
size=100000
dataf = pytdi.Data.from_gws( 'gw_galbin.h5', '../SMB_soon/orbits.h5', skipped=-size)
```

Pre-calculate the delays in the 'old' way, i.e. via Lagrange filters

```python
%%time
bb = X2.build(**dataf.args, parallel=False)
```

```python
%%time
sig = bb(dataf.measurements)
```

Switching on the new interpolation mode - time reduction should be about a factor 8 for the pre-calculation and about a factor 10 for the evaluation of the TDI signal.

```python
dataf_new = pytdi.Data.from_gws( 'gw_galbin.h5', '../SMB_soon/orbits.h5', interpolate=True, skipped=-size)
```

```python
%%time
bb_new= X2.build(**dataf_new.args, parallel=False)
```

```python
%%time
sig_new = bb_new(dataf_new.measurements)
```

## Checking the calulated signals

Plotting the 'old' way of calculating the signal against the 'new' way with the interpolator. Note that in both cases the ouput is an array. The points coincide very well, the difference is in the first few points where the 'old' method shows artefacts of the Lagrange filtering. UPDATE: The artefacts at the start of the signal have been significantly reduced by modifying the implementation of the lagrange filters a bit.


```python
fig, ax = plt.subplots( figsize=(9,6) )
ax.scatter(np.arange(0,100)/0.01,sig[:100], color='C0', marker='o', label="interpolate=False")
ax.plot(np.arange(0,100)/0.01,sig[:100], color='C0', linestyle='--')

ax.scatter(np.arange(0,100)/0.01,sig_new[:100], color='C1', marker='+',label="interpolate=True")
ax.plot(np.arange(0,100)/0.01,sig_new[:100],color='C1', linestyle=':')

ax.set_ylabel("Strain")
ax.set_xlabel("Time (s)")

ax.legend()
```

## Checking the delays and the delay derivatives
Caution! There is an additional minus sign between the 'new' delays and the 'old' delays (and the derivatives, respectively) - this is due to the way we calculate the shifted datastreams. This is usually hidden in the internals.

We plot every 20.000th point only, not to clog the graph. Dashed line give the interpolated results.

```python
fig, ax = plt.subplots( figsize=(9,6) )
dp = 20000
for ii, (k,v) in enumerate( dataf.delays.items()) :
    ax.scatter(dataf_new.t[0:-1:dp], v[0:-1:dp], color=f'C{ii}', label=k)
    ax.plot( dataf_new.t, -dataf_new.delays[k](dataf_new.t),color=f'C{ii}', linestyle='--')
ax.legend(bbox_to_anchor=(1.1, 1),
                         loc='upper left', borderaxespad=0.)

ax.set_xlabel('Time (s)')
ax.set_ylabel('Delay (s)')
secax = ax.secondary_yaxis( 'right', functions=( lambda x: -(-x*3e2+2500),  lambda x: -(-x*3e2+2500)))
secax.set_ylabel('Armlength wrt nominal (10$^3$ km)')
```

```python
fig, ax = plt.subplots( figsize=(9,6) )

for ii, (k,v) in enumerate( dataf.delay_derivatives.items()) :
    ax.scatter(dataf_new.t[0:-1:dp], v[0:-1:dp], color=f'C{ii}', label=k)
    ax.plot( dataf_new.t, -dataf_new.delay_derivatives[k](dataf_new.t),color=f'C{ii}', linestyle='--')
ax.legend(bbox_to_anchor=(1.1, 1),
                         loc='upper left', borderaxespad=0.)

ax.set_xlabel('Time (s)')
ax.set_ylabel('Delay derivative (s/s)')
secax = ax.secondary_yaxis( 'right', functions=( lambda x: -(-x*c),  lambda x: -(-x*c)))
secax.set_ylabel('Armlength change (m/s)')
```
